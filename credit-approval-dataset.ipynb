```python
import pandas as pd
import numpy as np
```

```python
#load the dataset
data = pd.read_csv("C:/Users/Vivek/Downloads/crx.data", header = None)
```

```python
#Variable Engineering
#col1 col2 .... col16
varnames = ['col' + str(s) for s in range(1,17)]
```

```python
#add column names
data.columns = varnames
```

```python
data.tail(10) # last 10 rows of data
```

```python
# nan -> not a numeric value
#replace ? with np.nan

data = data.replace('?', np.nan)
```

```python
data.info()

#includes information about datatypes of variables 
#in pandas object -> string
```

```python
# recast some variables to their correct datatypes
data['col2'] = data['col2'].astype('float')
data['col14'] = data['col14'].astype('float')
```

```python
data.info()
```

```python
# encode col16 values from +/- to p/n
# encoding
```

```python
data['col16'] = data['col16'].map({'+':'p', '-':'n'})
```
